```python
# Import the packages for this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# Read the data
df = pd.read_csv(r'C:\Users\lkhoo\OneDrive\Documents\Python Scripts\movies.csv')
```

```python
df
```

```python
# Check to see if there's missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#For purpose of this project, delete rows with missing data

df = df.dropna()

df
```

```python
# Check data types in columns

print(df.dtypes)
```

```python
# Change data type in columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# Create new column for correct year (yearcolumn)

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
# Drop duplicates, if any

df.drop_duplicates()
```

```python
# Sort data by Top Grossing Film

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# After cleaning data, looking at some assumptions for data visualization.
```

```python
# Assumption 1. Budget of film correlation with Gross Revenue.

# Scatterplot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Using seaborn, plot budget vs gross

sns.regplot(x="gross", y="budget", data=df, scatter_kws={'color' : "blue"}, line_kws={'color' : 'red'})
```

```python
# Create correlation matrix

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Factors")

plt.xlabel("Movie information")

plt.ylabel("Movie information")

plt.show()
```

```python
# Looking at Company

df.head()
```

```python
# Create numeric expression for Data type - objects
# Create unique identifier for Company (INT)

df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# Create correlation matrix

correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie information")

plt.ylabel("Movie information")

plt.show()
```

```python
# Visualize matrix

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Pairs with high correlation (>0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

strong_pairs
```
